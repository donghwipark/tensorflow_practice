```python
import requests
from bs4 import BeautifulSoup
res = requests.get('https://finance.naver.com/')
soup = BeautifulSoup(res.content,'html.parser')
section = soup.select('div.aside_area.aside_stock > table > tbody > tr')
for item in section:
    print("지수 이름: ", item.find('a').get_text().strip(), "지수: ",item.find('td').get_text(), "지수변동: ", item.find('span').get_text())
```

```python
import openpyxl
import requests
from bs4 import BeautifulSoup
res = requests.get('https://finance.naver.com/')
soup = BeautifulSoup(res.content,'html.parser')
section = soup.select('div.aside_area.aside_stock > table > tbody > tr')

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active

for index in range(9):
    excel_sheet.append([index, 'hello'])
for item in section:
    excel_sheet.append(["지수 이름: " + item.find('a').get_text().strip(), "지수: " + item.find('td').get_text(), "지수변동: " + item.find('span').get_text()])
    
excel_file.save('test.xlsx')
excel_file.close()
```
